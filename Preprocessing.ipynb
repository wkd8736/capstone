```python
file_path = "/content/gdrive/My Drive/Dataset2.csv"  
data = pd.read_csv(file_path, sep='\t')
```

```python
data.info()
```

```python
# 인식하지 못하는 행 삭제

row_5688 = data.loc[5687]  # 인덱스는 0부터 시작하므로 5689번 행의 인덱스는 5688입니다.
print(row_5688)

row_5687 = data.loc[10036]  # 인덱스는 0부터 시작하므로 5689번 행의 인덱스는 5688입니다.
print(row_5687)

row_5686 = data.loc[12369]  # 인덱스는 0부터 시작하므로 5689번 행의 인덱스는 5688입니다.
print(row_5686)

rows_to_delete = [5687, 5688, 12369]  # 삭제할 행의 인덱스

data = data.drop(rows_to_delete)
```

```python
data.info()
```

```python
# 결측값이 있는 행 삭제
null = data[data.lable.isnull()].index
data.loc[null,"content"]
```

```python
data = data.dropna(subset=['lable'])
```

```python
data = data.astype({"lable":"int"})
data.info()
```

```python
print(data)
```

```python
data.to_csv("/content/gdrive/My Drive/last_Dataset.csv", index=False)
```
